## Book to plot fire coordinates in Python using the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Fires.xlsx')
latitude = df['POINT_X']
longitude = df['POINT_Y']

plt.figure(figsize=(10, 6))
plt.scatter(longitude, latitude, marker='o', color='b')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Lat-Lon Scatter Plot')
plt.grid(True)

# Set axis limits
plt.xlim(min(longitude) - 1, max(longitude) + 1)
plt.ylim(min(latitude) - 1, max(latitude) + 1)

# Display the plot in the console
plt.show()

# Display the plot in the console
plt.show()
```
