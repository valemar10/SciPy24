### Visualize an image of a forested area before a wildfire

```python
# File path to a 6 channel GeoTIFF input
file_path = '/Volumes/HD710PRO/Fire_and_Hurricane_Images/Fire/6channel/Training/Damage/RGB_6Channel646_tile_104_vert.tif'

# Read the image using tifffile
rgb_image = tifffile.imread(file_path)

# Select the first three channels to visualize the forest before a wildfire
selected_channels = rgb_image[..., :3]  # Selecciona los primeros tres canales

# Visualizar la imagen con los canales seleccionados
tifffile.imshow(selected_channels)
```

### Visualize an image of the same forested area after a wildfire
#### Notice the change in the greenness

```python
# Select the last three channels for visualization
last_three_channels = rgb_image[..., -3:]  # Select the last three channels

# Visualize the image with the last three channels
tifffile.imshow(last_three_channels)
```
